### Descrizione del dataset
Nel dataset sono presenti 100 diverse istanze di problema, ottenute dall'osservazione di un Virtual Power Plant (VPP). Su ciascuna è stato fatto girare per 100 volte l'algoritmo online di fixing considerando ogni volta un numero diverso di tracce (da 1 a 100).  
Quindi **il dataset è formato da 100run x 100istanze = 10000 rows**.

Gli attributi **Load(kW) e PV(kW)** sono dei vettori relativi alle osservazioni di carico e di quantità di energia fotovoltaica disponibile. Ogni vettore contiene 96 valori campionati online in 96 stage (ogni 15 minuti nell'arco di una giornata, quindi 4 campionamenti all'ora x 24 ore = 96 stage). 

Nelle prime 100 righe del dataset avremo i risultati dell'algoritmo di fixing fatto girare sulle 100 diverse istanze ma considerando 1 sola traccia. Poi nelle seconde 100 rows i risultati dell'algoritmo di fixing applicato sempre a ciascuna di quelle 100 istanze ma stavolta considerando 2 tracce... ecc.
Gli attributi **sol(keuro) e time(sec)** sono rispettivamente i valori della soluzione e del tempo necessario al runtime online dell'algoritmo e sono stati ottenuti come somma su tutto l'orizzonte temporale (sui 96 stage) di tutte le 96 soluzioni e tempi di running parziali.  
L'attributo **memAvg(MB)** rappresenta la memoria utilizzata dall'algoritmo sulla macchina su cui è stato eseguito. E' una media della memoria usata sui 96 stage.

## Modelli Predittivi

L'obiettivo di questa sezione è quello di sviluppare dei modelli predittivi di Machine Learning che siano in grado di effettuare delle predizioni riguardo al valore di un attributo, che chiameremo *target*, sulla base dei valori assunti da altri attributi, detti *feature*.

Verranno indagati alcuni modelli regressori e comparati tra loro sfruttando delle opportune metriche.

### Predire *nTraces* conoscendo le feature *sol(keuro)*, *time(sec)* e *memAvg(MB)*

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor  
from sklearn.model_selection import train_test_split
from matplotlib.lines import Line2D
from sklearn.tree import export_graphviz
```

#### Linear Regression

```python
## FUNZIONI PER LA STAMPA DEI GRAFICI DI CONFRONTO METRICHE ##
def autolabel(rects, ax):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{:.3f}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 1),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')
def plot_scores_comparison(R2_scores, MSE_scores):
    fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True)
    fig.set_size_inches(7,10)
    plt.xticks(rotation=-80)
    rects1 = ax1.bar(R2_scores.keys(), R2_scores.values(), label="R2")
    ax1.legend()
    rects2 = ax2.bar(MSE_scores.keys(), MSE_scores.values(), label="MSE", color="red")
    ax2.legend()
    autolabel(rects1, ax1)
    autolabel(rects2, ax2)
    fig.tight_layout()
    plt.show()
```

```python
# Divido il dataset in feature e target.
# NOTA: le feature sol(keuro), time(sec) e memAvg(MB) verranno utilizzate in diverse combinazioni per allenare il modello.

X1 = cleandf.loc[:, ['sol(keuro)', 'time(sec)', 'memAvg(MB)']]  #usa questo per la multiple regression
X2 = cleandf.loc[:, ['sol(keuro)', 'time(sec)']]                #usa questo per la multiple regression
X3 = cleandf.loc[:, ['sol(keuro)', 'memAvg(MB)']]               #usa questo per la multiple regression
X4 = cleandf.loc[:, ['time(sec)', 'memAvg(MB)']]                #usa questo per la multiple regression
X5 = cleandf.loc[:, ['sol(keuro)']]                             #usa questo per la linear regression con solo la feature sol(keuro)
X6 = cleandf.loc[:, ['time(sec)']]                              #usa questo per la linear regression con solo la feature time(sec)
X7 = cleandf.loc[:, ['memAvg(MB)']]                             #usa questo per la linear regression con solo la feature memAvg(MB)
y  = cleandf.loc[:, ['nTraces']]                                #target to predict

X_set = [X1, X2, X3, X4, X5, X6, X7]
linear_R2_scores  = {}
linear_MSE_scores = {}

for X in X_set :
    print()
    print("---------- MODELLO DI LINEAR REGRESSION ALLENATO CONSIDERANDO LE FEATURE: " + str(X.columns.values) + " ----------")
    # Faccio uno split random in train set e test set.
    # NOTA: Per evitare overfitting del modello e considerando che disponiamo di un cospicuo numero di record,
    #       la dimensione del train set non è eccessivamente alta ma è abbastanza equilibrata rispetto alla dimensione del test set.
    #       train set: 55% - test set: 45%.
    Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, test_size=0.45, random_state=1)

    # Fit del modello
    reg = LinearRegression().fit(Xtrain, ytrain)

    Xtest = Xtest.sort_index()
    ytest = ytest.sort_index()

    # prediction on test set
    nTraces_predicted_linear = reg.predict(Xtest)
    
    ######### VISUALIZZAZIONE DELLE PREDIZIONI #########
    plt.scatter(nTraces_predicted_linear, ytest['nTraces'], s=0.5, color='blue', alpha = 0.2)
    plt.plot(nTraces_predicted_linear, nTraces_predicted_linear, color='red')
    # specify title 
    plt.title('Linear Regression prediction')  
    #specify X axis label 
    plt.xlabel('') 
    plt.xticks(()) 
    #specify Y axis label 
    plt.ylabel('')
    plt.yticks(())
    #plt.yticks(np.arange(1, 120, step = 10))
    #legend
    custom_lines = [Line2D([0], [0], color='blue'),
                    Line2D([0], [0], color='red')]
    legend = plt.legend(custom_lines, ['true value', 'predicted value'], loc="upper center")
    plt.show()

    ######### CALCOLO DELLE METRICHE #########
    # Calcolo la metrica "R_squared" (R^2)
    r2 = r2_score(ytest, nTraces_predicted_linear)
    print('Score R2:'+'\n'+ str(r2))    
    linear_R2_scores[str(X.columns.values)] = r2 
    # Calcolo la metrica MSE
    mse = mean_squared_error(ytest, nTraces_predicted_linear)
    linear_MSE_scores[str(X.columns.values)] = mse
    print('Mean Squared Error (MSE):'+'\n'+ str(mse))
    print()
    
    
plot_scores_comparison(linear_R2_scores, linear_MSE_scores)  
```

**COMMENTO**  
I risultati ottenuti rispecchiano le considerazioni sulle *correlation Matrix* fatte in sezione 2.1. 

Le feature maggiormente correlate con nTraces sono *time(sec)* e *memAvg(MB)*. Non ci sorprende dunque il fatto che i risultati di predizione migliori sono stati ottenuti quando il modello viene allenato considerandole entrambe.
Al secondo posto nella classifica delle performance troviamo i risultati di predizione ottenuti considerando *time(sec)* ed al terzo quelli ottenuti considerando *memAvg(MB)*.
In questa classifica non è stata volutamente citata *sol(keuro)* in quanto possiamo notare che la sua presenza tra le feature su cui allenare il modello è pressochè irrilevante in termini di performance.  

*Esempio:* La R2 sul modello di multiple regression in cui gli attributi usati per allenare il modello sono memAvg(MB), time(sec) e sol(keuro) è la migliore ottenuta (96.1%), ed è pressochè identica alla R2 ottenuta senza sol(keuro) (96%). Anche l'MSE senza sol(keuro) rimane quasi invariato (peggiora di 0.3).

Questo comportamento è giustificato dal fatto che **sol(keuro) è scarsamente correlata con nTraces** e quindi **non è molto utile alla predizione del target**. La R2 sul modello di linear regression in cui considero solo sol(keuro) infatti è parecchio scarsa (circa 4%).

#### Regression Tree

```python
tree_R2_scores  = {}
tree_MSE_scores = {}
for X in X_set :
    print("---------- MODELLO DI REGRESSION TREE ALLENATO CONSIDERANDO LE FEATURE: " + str(X.columns.values) + " ----------")
    # Faccio uno split random in train set e test set.
    # NOTA: Per evitare overfitting del modello e considerando che disponiamo di un cospicuo numero di record,
    #       la dimensione del train set non è eccessivamente alta ma è abbastanza equilibrata rispetto alla dimensione del test set.
    #       train set: 55% - test set: 45%.
    Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, test_size=0.45, random_state=1)

    # create a regressor object 
    regressor_tree = DecisionTreeRegressor(max_depth = 10, random_state = 0)  

    # fit the regressor with X and Y train dataset  
    regressor_tree.fit(Xtrain, ytrain)
    
    Xtest = Xtest.sort_index()
    ytest = ytest.sort_index()
    
    # prediction on test set
    nTraces_predicted_tree = regressor_tree.predict(Xtest)
    
    ######### VISUALIZZAZIONE DELLE PREDIZIONI #########
    plt.scatter(nTraces_predicted_tree, ytest['nTraces'], s=0.5, color='blue', alpha = 0.3)
    plt.plot(nTraces_predicted_tree, nTraces_predicted_tree, color='red')
    # specify title 
    plt.title('Regression Tree prediction')  
    #specify X axis label 
    plt.xlabel('') 
    plt.xticks(()) 
    #specify Y axis label 
    plt.ylabel('')
    plt.yticks(())
    #plt.yticks(np.arange(1, 120, step = 10))
    #legend
    custom_lines = [Line2D([0], [0], color='blue'),
                    Line2D([0], [0], color='red')]
    legend = plt.legend(custom_lines, ['true value', 'predicted value'], loc="upper center")

    plt.show()

    ######### EXPORT DEL TREE IN UN FILE #########
    #Store the regression tree in a tree.dot file in order to visualize the plot.
    #Visualize it on http://www.webgraphviz.com/ by copying and pasting relating data from the .dot file
    
    #export_graphviz(regressor_tree, out_file = "./RegressorTreeGraph/RegressorTreeGraph"+str(X.columns.values)+".dot", feature_names=X.columns.values)
    
    
    ######### CALCOLO DELLE METRICHE #########
    # Calcolo la metrica "R_squared" (R^2)
    r2 = r2_score(ytest, nTraces_predicted_tree)
    print('Score R2:'+'\n'+ str(r2))    
    tree_R2_scores[str(X.columns.values)] = r2 
    # Calcolo la metrica MSE
    mse = mean_squared_error(ytest, nTraces_predicted_tree)
    tree_MSE_scores[str(X.columns.values)] = mse
    print('Mean Squared Error (MSE):'+'\n'+ str(mse))
    print()
    
plot_scores_comparison(tree_R2_scores, tree_MSE_scores)
```

- I risultati ottenuti con i modelli di Regression Tree sono migliori di quelli ottenuti dalla Linear Regression, in quanto come visualizzabile dai grafici in 2.2, le feature non presentano tra loro delle relazioni lineari.  

- Parecchi modelli qui allenati hanno un R2 prossimo a 1 e un MSE prossimo a 0.

- Sebbene la R2 della sol(keuro) sia migliorata (ora è al 42%) vale ancora la stessa osservazione fatta per la linear Regression:  **la sua presenza/assenza non inficia significativamente sulle performance predittive ottenute con le altre feature.**

#### Rafforzare il peso della feature *sol(keuro)* utilizzando le feature *PV(kW)* e *load(kW)*

Come evinto dalle considerazioni precedenti, per predirre in modo accurato il numero di Tracce le feature più convenienti da considerare sono *time* e *memAvg* rispetto alla feature *sol*.  
Lo scenario di utilità reale che si prospetta quindi è: "*abbiamo un certo tempo e/o una certa quantità di memoria a disposizione e dobbiamo far eseguire l'algoritmo di fixing, quante tracce possiamo fargli prendere in considerazione?*" 

Tuttavia nella pratica si potrebbero avere esigenze anche diverse. Ad esempio: "*voglio ottenere una certa qualità di soluzione. Quante tracce devo far prendere in considerazione all'algoritmo?*"  
Visto quanto realizzato fino ad ora, se si considera un modello predittivo che considera solo la feature soluzione, non si ottengono predizioni su *nTraces* molto accurate.

Lo scopo di questo paragrafo è quello di **considerare insieme alla sol(keuro) anche le feature *PV(kW)* e *load(kW)***, da cui essa è fortemente dipendente, **ai fini di sviluppare un modello predittivo con target *nTraces* che sia più accurato.**

**Considerazioni su cui ragionare**: PV e load sono dei vettori di 96 elementi per ogni istanza, mentre sol(keuro) è un unico valore ottenuto come somma delle 96 soluzioni sui 96 stage. Quindi come è possibile trattarli? Con una media? O altro?

Al momento sono stati trattati considerandone la media aritmetica.

```python
# X5 = cleandf.loc[cleandf['sol(keuro)'].astype(float)<=upper_quartile+1.5*iqr, ['sol(keuro)']]      #usa questo per la linear regression con solo la feature sol(keuro)
# y  = cleandf.loc[cleandf['sol(keuro)'].astype(float)<=upper_quartile+1.5*iqr, ['nTraces']]   
mean_pv   = []
mean_load = []

for index, value in cleandf.loc[0:99, 'PV(kW)'].items() :
    pv = str(value)
    pv = pv.replace('[', "").replace(']', "").replace("\\", "").replace("n","").replace('\'', "") #pulizia necessaria per poi fare il cast a float
    pv = str.split(pv)
    pv = np.array(pv).astype(float)
    mean_pv.append(np.mean(pv))
        
for index, value in cleandf.loc[0:99, 'Load(kW)'].items() :
    load = str(value)
    load = load.replace('[', "").replace(']', "").replace("\\", "").replace("n","").replace('\'', "") #pulizia necessaria per poi fare il cast a float
    load = str.split(load)
    load = np.array(load).astype(float)
    mean_load.append(np.mean(load))
# Per ovvi motivi di efficienza, vista la periodicità dei valori di PV e Load, anzichè calcolarli tutti e 10000 abbiamo calcolato solo i primi 100. Ora li replichiamo.
for i in range(1,100):
    for j in range(0,100):
        mean_pv.append(mean_pv[j])
        mean_load.append(mean_load[j])
```

## Neural Network: Multi-Layer Perceptron regressor

L'obiettivo di questa sezione è quello di sviluppare una Neural Network in grado di predire il valore di un attributo target, dati in input i valori di altri 3 attributi feature.

A questo fine sono stati considerati gli attributi [*'nTraces', 'sol(keuro)', 'time(sec)', 'memAvg(MB)'*] dei quali uno, a rotazione, viene considerato come target. 

Il modello di Neural Network che utilizzerò è il [Multi-Layer Perceptron regressor.](https://scikit-learn.org/stable/modules/generated/sklearn.neural_network.MLPRegressor.html#sklearn.neural_network.MLPRegressor)

### Feature scaling (Normalizzazione del dataset)

La rete neurale verrà allenata su 3 differenti Dataset:
- dataset con i valori originali.
- dataset con i valori standardizzati (normalizzazione Z-score).
- dataset con i valori normalizzati (normalizzazione Min-Max scaling).

La [normalizzazione Z-Score:](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.StandardScaler.html)
    - viene calcolata come z = (x - u) / s  
      dove u è la media dei campioni, e s è la deviazione standard.
    - non produce dati in un range unitario ma li riconduce ad una distribuzione di media = 0 e deviazione std = 1.

La [normalizzazione min-max scaling:](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.MinMaxScaler.html)
    - viene calcolata come z = (x - min(x)) / (max(x) - min(x))
    - produce dati in un range unitario [0,1] .
    

Queste tecniche di **feature scaling** (ridimensionamento dei dati) sono molto utili per uniformare i valori delle varie feature **evitando in questo modo che una feature possa dominare sulle altre** ed incidere maggiormente sul training della rete solo perchè presenta valori più grandi.

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler

standardizer = StandardScaler()
minMaxNormalizer = MinMaxScaler()

df_original = cleandf.loc[:, ['nTraces', 'sol(keuro)', 'time(sec)', 'memAvg(MB)']]
df_original = df_original.astype(float)
# df_original.plot.kde()

# NORMALIZZAZIONE VALOR MEDIO E DEV STD (STANDARDIZZAZIONE)
standardizedData = standardizer.fit_transform(df_original)
df_standardized = pd.DataFrame(data=standardizedData, columns=['nTraces', 'sol(keuro)', 'time(sec)', 'memAvg(MB)'])
# df_standardized.plot.kde()

# NORMALIZZAZIONE MIN MAX SCALING
normalizedMinMaxData = minMaxNormalizer.fit_transform(df_original)
df_normalizedMinMax  = pd.DataFrame(data=normalizedMinMaxData, columns=['nTraces', 'sol(keuro)', 'time(sec)', 'memAvg(MB)'])
# df_normalizedMinMax.plot.kde()
```

### NN (Simple Holdout Method)
In questa sezione gli MLP regressor vengono allenati con il metodo **Simple Holdout**, ossia dividendo il dataset a disposizione in Train Set + Test Set.  
Il modello viene allenato sul Train Set e poi valutato sul Test Set.  
In questa sezione sono stati usati i parametri di default del MLP regressor.  

```python
from sklearn.neural_network import MLPRegressor
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import cross_val_predict

df_to_test = {
    "original"     : df_original,
    "standardized" : df_standardized,
    "normalized"   : df_normalizedMinMax
};

target_list = ['nTraces', 'sol(keuro)', 'time(sec)', 'memAvg(MB)']

for key, df in df_to_test.items():  # Per ogni dataframe da testare...
    print("__________________________________________________________________________________________")
    printmd("- Testing the Dataframe with **" + key + "** values")
    for target in target_list:  # ...a rotazione cambio il target da predirre
        printmd("Target to predict: **" + target + "**")
        X = df.drop(target, axis=1)
        y = df[target]
        #SIMPLE HOLDOUT METHOD (TRAIN-TEST)
        Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, test_size=0.45, random_state=1)
        mlp_regr = MLPRegressor(random_state=1).fit(Xtrain, ytrain)
        printmd("R2 score: " + str(mlp_regr.score(Xtest, ytest)) )  
    
```

**COMMENTO:**
- I risultati migliori sono stati ottenuti dal MLP allenato sul dataset con i valori standardizzati. A seguire MLP allenato sul dataset con i valori normalizzati ed infine MLP allenato sul dataset con i valori originali.
- Come già osservato nei paragrafi precedenti, i target che possono essere predetti con maggior precisione sono *nTraces* e *time*.
- La predizione del target sol(keuro) restituisce ancora una volta scarse prestazioni.

### Tuning the Hyperparameters with GridSearchCV
In questa sezione viene fatto il tuning di alcuni parametri del modello MLP con l'obiettivo di trovare quelli che garantiscono le migliori performance predittive (secondo la scoring function R2).

La  [GridSearchCV](https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.GridSearchCV.html) è una classe Python fornita dalla libreria sklearn che permette di eseguire, per un modello predittore, una ricerca esaustiva tra tutte le possibili combinazioni dei parametri forniti (una cosiddetta ParameterGrid).  Quando viene fatta la fit sul dataset, infatti, la GridSearch valuta le performance dei modelli ottenuti con tutte le possibili combinazioni di parametri. La combinazione migliore viene poi scelta per creare il modello finale su cui poter invocare la funzione *predict*.

I parametri che vengono testati sono:
- **activation**: è l'activation function utilizzata negli hidden layer.
- **hidden_layer_sizes'**: Il numero di hidden layer e, per ciascuno, la sua dimensione in numero di neuroni. A default si ha 1 hidden layer di 100 neuroni.
- **max_iteration**: numero massimo di iterazioni a cui, anche se non si è raggiunta un'ottimizzazione, il training della rete si stoppa.
- **tol**: fattore di tolleranza per l'ottimizzazione. Se per più di *n_iter_no_change (default = 10)* iterazioni consecutive la scoring function non migliora di almeno *tol*, l'ottimizzazione si considera raggiunta e il training della rete si stoppa.


La fit sul dataset viene effettuata sfruttando un **KFold Cross-Validation** (con K=5). Il dataset cioè viene diviso in 5 parti (Fold) e viene eseguito il Train su K-1 di questi e il test sul restante. La procedura viene eseguita K volte cambiando ogni volta il Fold designato come Test Set. Infine viene fatta una media della metrica calcolata ad ogni iterazione.

A fronte di un maggior costo computazionale, la Cross Validation porta i seguenti vantaggi:
 - permette di allenare il modello utilizzando tutti i dati del dataset. 
 - permette di ottenere una stima della performance che, essendo basata su una media di K iterazioni, è più affidabile.
 - tutte le istanze del dataset vengono usate una volta per il testing.


```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import KFold

mlp_parameters = {
    'activation'         : ['identity', 'logistic', 'tanh', 'relu'],
    'hidden_layer_sizes' : [(100,), (100, 100), (100, 100, 100), (100, 100, 100, 100)],   
    'max_iter'           : list(100*i for i in range(1, 6)),                             # 100, 200, 300, 400, 500
    'tol'                : list(10**i for i in range(-4, -7, -1))                        # 1e-4, 1e-5, 1e-6 
}
    
for key, df in df_to_test.items():  # Per ogni dataframe da testare...
    print("__________________________________________________________________________________________")
    printmd("- Testing the Dataframe with **" + key + "** values")
    for target in target_list:  # ...a rotazione cambio il target da predirre
        printmd("Target to predict: **" + target + "**")
        X = df.drop(target, axis=1)
        y = df[target]
        
        # Voglio che la CV faccia uno shuffle dei dati per ottenere dei Fold con sample casuali. Altrimenti, essendo il dataset ordinato per nTraces, avrei dei fold che presentano solo istanze con uno (o alcuni) valori di *nTraces* e questo inficierebbe sulle performance.
        cv = KFold(n_splits=5, shuffle=True, random_state=1)

        #GRID SEARCH CV METHOD
        mlp_regr = MLPRegressor(random_state=1)
        clf = GridSearchCV(mlp_regr, mlp_parameters, cv=cv, scoring='r2', n_jobs = -1) # this allows to parallelize computations using all the available cores
        clf.fit(X, y)
        print("Best parameters set found :")
        print(clf.best_params_)
        print()
        print("Best R2 score found :")
        print(clf.best_score_)
        print()
```

**COMMENTO** 

I risultati ottenuti dai modelli MLP con la miglior combinazione di parametri trovata dalla GridSearchCV evidenziano che:
- le performance generali sono molto buone (sempre eccezion fatta per i modelli con target = sol) ma sono solo leggermente migliori delle performance ottenute in sezione 4.2 dal MLP regressor allenato con i parametri di default.
- l'unica prestazione che è migliorata significativamente è la predizione di memAvg(MB) nel dataset con i valori standardizzati. Lo score R2 migliora di circa 17 punti percentuali arrivando a segnare un punteggio di 91.5%.
